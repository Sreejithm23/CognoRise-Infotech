```python
df['Brand']=df['CarName'].apply(lambda x:x.split()[0])
df
```

```python
df.drop(['CarName'],axis=1,inplace=True)
df
```

```python
df.dtypes
```

```python
df1=pd.get_dummies(df[['fueltype','aspiration','doornumber','carbody','drivewheel','enginelocation','enginetype','cylindernumber','fuelsystem','Brand']],dtype=int,drop_first=True)
df1
```

```python
dfe=pd.concat([df,df1],axis=1)
dfe
```

```python
dfe.drop(['fueltype','aspiration','doornumber','carbody','drivewheel','enginelocation','enginetype','cylindernumber','fuelsystem','Brand'],axis=1,inplace=True)
dfe
```

```python
dfe.columns
```

```python
dfe.dtypes
```

```python
x=dfe.drop(['price'],axis=1)
x
```

```python
y=dfe['price']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
df2=pd.DataFrame({'Actual Value':y_test,'Predicted Value':y_pred,'Difference':y_test-y_pred})
df2
```

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error,r2_score
print("MAE is",mean_absolute_error(y_test,y_pred))
print("MSE is",mean_squared_error(y_test,y_pred))
mse=mean_squared_error(y_test,y_pred)
print("RMSE is",np.sqrt(mse))
print("MAPE is",mean_absolute_percentage_error(y_test,y_pred))
print("R2 Score is",r2_score(y_test,y_pred))
```
